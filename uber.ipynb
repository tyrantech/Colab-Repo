```python
import pandas as pd
import sqlite3
df = pd.read_csv("ncr_ride_bookings.csv")
conn = sqlite3.connect(":memory:")
df.to_sql('rides',conn,index=False,if_exists="replace")
pd.read_sql("SELECT * FROM rides",conn)
```

```python
df.head()
```

```python
#sum fares per ride type
pd.read_sql("""SELECT "Vehicle Type","Driver Ratings",SUM("Booking Value") AS highest_cost
FROM rides
GROUP BY "Vehicle Type","Driver Ratings"
ORDER BY highest_cost DESC""",conn)
```

```python
import matplotlib.pyplot as plt
df.rename(columns={df.columns[1]: "highest_cost"}, inplace=True)
plt.bar(df["Vehicle Type"], df["highest_cost"], color="skyblue")
plt.xlabel("Vehicle Type")
plt.ylabel("Total Revenue")
plt.title("Revenue by Vehicle Type")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.read_sql("""SELECT Date,SUBSTR(Time,1,2) AS Hour,SUM("Booking Value") AS
prices
FROM rides
GROUP BY hour
ORDER BY prices DESC""",conn)
```
