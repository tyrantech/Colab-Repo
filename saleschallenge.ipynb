```python
import pandas as pd
import sqlite3
sales = [
    {"product": "apple", "amount": 10, "category": "fruit"},
    {"product": "banana", "amount": 5, "category": "fruit"},
    {"product": "carrot", "amount": 7, "category": "vegetable"},
    {"product": "spinach", "amount": 3, "category": "vegetable"},
    {"product": "orange", "amount": 8, "category": "fruit"},
    {"product": "almond", "amount": 15, "category": "nut"}
]

#convert to dataframe
df = pd.DataFrame(sales)

#create sqlite in-memory database
conn = sqlite3.connect(':memory:')
df.to_sql('sales',conn,index=False,if_exists='replace')

pd.read_sql("SELECT * FROM sales",conn)
```

```python
#calculate total amount sold per category
sold = pd.read_sql("""SELECT category,SUM(amount) AS total_per_cat
FROM sales
GROUP BY category;""",conn)
print(sold)
total_amount_of_fruit = sold.loc[sold['category']=='fruit','total_per_cat'].values[0]
total_amount_of_vegetable = sold.loc[sold['category']=='vegetable','total_per_cat'].values[0]

print("fruit:",total_amount_of_fruit)
print("vegetable:",total_amount_of_vegetable)
```

```python
#3 category with the most sold
sold_categories= pd.read_sql("""SELECT category,SUM(amount) AS total_per_cat
FROM sales
GROUP BY category;""",conn)
maxTotal = sold_categories['total_per_cat'].max()
mostSold = sold_categories[sold_categories['total_per_cat']== maxTotal]
print(mostSold)
```

```python
minTotal = sold_categories['total_per_cat'].min()
minSold = sold_categories[sold_categories['total_per_cat']== minTotal]
print(minSold)
```

```python
totals = {}  # start with an empty dictionary

for sale in sales:
    category = sale['category']  # get the category, like 'fruit'
    amount = sale['amount']      # get the amount, like 10

    # Check if category is not in totals dictionary
    if category not in totals:
        totals[category] = 0    # if new category, start count at 0

    # Add current sale amount to that category's total
    totals[category] = totals[category] + amount

print(totals)  # show the final totals dictionary
```
