```python
import sqlite3
import pandas as pd

# Connect to in-memory database
conn = sqlite3.connect(':memory:')

# Create sales table
conn.execute('''
CREATE TABLE sales (
    sale_id INTEGER PRIMARY KEY,
    customer TEXT,
    product TEXT,
    category TEXT,
    quantity INTEGER,
    price REAL,
    sale_date TEXT
)
''')

# Insert sample data
sales_data = [
    (1, 'Alice', 'Apple', 'Fruit', 5, 10, '2025-08-01'),
    (2, 'Alice', 'Chips', 'Snack', 2, 15, '2025-08-02'),
    (3, 'Bob', 'Banana', 'Fruit', 10, 5, '2025-08-01'),
    (4, 'Bob', 'Carrot', 'Vegetable', 3, 8, '2025-08-03'),
    (5, 'Charlie', 'Broccoli', 'Vegetable', 4, 12, '2025-08-02'),
    (6, 'Charlie', 'Chocolate', 'Snack', 1, 20, '2025-08-04'),
    (7, 'Diana', 'Apple', 'Fruit', 2, 10, '2025-08-03'),
    (8, 'Diana', 'Carrot', 'Vegetable', 5, 8, '2025-08-04'),
    (9, 'Diana', 'Chips', 'Snack', 3, 15, '2025-08-04')
]

conn.executemany('INSERT INTO sales VALUES (?, ?, ?, ?, ?, ?, ?)', sales_data)
conn.commit()
pd.read_sql("SELECT * FROM sales",conn)
```

```python
#quantity > 3
pd.read_sql("SELECT * FROM sales WHERE quantity > 3",conn)
```

```python
#2
pd.read_sql("""SELECT * from SALES WHERE category ='Fruit' AND price > 5""",conn)
```

```python
#total sales per category
pd.read_sql("""SELECT category,SUM(quantity * price) AS sales_per_category
FROM sales
GROUP BY category
""",conn)
```

```python
#average price per category
sold = pd.read_sql("""SELECT category,price FROM sales""",conn)
average = sold['price'].mean()
print("average is",average)
```

```python
#date filtering
pd.read_sql("""SELECT customer,sale_date FROM sales
WHERE sale_date BETWEEN '2025-08-02' AND '2025-08-04'""",conn)
```

```python
#combined challenge
pd.read_sql("""SELECT customer,MAX(quantity * price) as the_most
FROM sales
GROUP BY customer
""",conn)
```

```python
#topselling
pd.read_sql("""SELECT product,quantity
FROM sales
WHERE quantity = (SELECT MAX(quantity) FROM sales)""",conn)
```
